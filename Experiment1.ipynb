```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('DMart.csv')
```

```python
df.head()
df.info()
df.shape
```

```python
df.isna().sum()
df['Category'].fillna(df['Category'].mode()[0], inplace=True)
df['Quantity'].fillna(df['Quantity'].mode()[0], inplace=True)
df['Brand'].fillna(df['Brand'].mode()[0], inplace=True)
df['Price'].fillna(df['Price'].mode()[0], inplace=True)
df['DiscountedPrice'].fillna(df['DiscountedPrice'].mode()[0], inplace=True)
df['SubCategory'].fillna(df['SubCategory'].mode()[0], inplace=True)
df['Description'].fillna(df['Description'].mode()[0], inplace=True)
df['BreadCrumbs'].fillna(df['BreadCrumbs'].mode()[0], inplace=True)
df['Name'].fillna(df['Name'].mode()[0], inplace=True)
print(df.isna().sum())
```

```python
df['Category'].value_counts()
```

```python
# Calculate discount percentage for a clearer view of discounts
df['DiscountPercentage'] = ((df['Price'] - df['DiscountedPrice']) / df['Price']) * 100

# 1. Top Categories by Product Count
top_categories = df['Category'].value_counts().head(5)
plt.figure(figsize=(8, 5))
sns.barplot(x=top_categories.index, y=top_categories.values, palette='Set2')
plt.title('Top Categories by Product Count')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

# 2. Top Brands by Product Count
top_brands = df['Brand'].value_counts().head(5)
plt.figure(figsize=(8, 5))
sns.barplot(x=top_brands.index, y=top_brands.values, palette='Set3')
plt.title('Top Brands by Product Count')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()

# 3. Average Price by Subcategory (only top 5 subcategories by count for clarity)
top_subcategories = df['SubCategory'].value_counts().head(5).index
avg_price_subcat = df[df['SubCategory'].isin(top_subcategories)].groupby('SubCategory')['Price'].mean()
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_subcat.index, y=avg_price_subcat.values, palette='viridis')
plt.title('Average Price by Top SubCategories')
plt.xlabel('SubCategory')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()

# 4. Discount Percentage Distribution for Top Brands
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[df['Brand'].isin(top_brands.index)], x='Brand', y='DiscountPercentage', palette='coolwarm')
plt.title('Discount Percentage by Top Brands')
plt.xlabel('Brand')
plt.ylabel('Discount Percentage')
plt.show()

# 5. Quantity Distribution (Filter common quantities for clarity)
common_quantities = df['Quantity'].value_counts().head(5).index
plt.figure(figsize=(8, 5))
sns.countplot(data=df[df['Quantity'].isin(common_quantities)], x='Quantity', palette='magma')
plt.title('Distribution of Common Quantities')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.show()

# 6. Price Comparison between Original and Discounted Price for Top 5 Brands
plt.figure(figsize=(10, 6))
sns.barplot(data=df[df['Brand'].isin(top_brands.index)], x='Brand', y='Price', color='blue', alpha=0.6, label='Original Price')
sns.barplot(data=df[df['Brand'].isin(top_brands.index)], x='Brand', y='DiscountedPrice', color='red', alpha=0.6, label='Discounted Price')
plt.title('Price Comparison between Original and Discounted Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.legend()
plt.show()

# Observations and Insights
print("Observations and Insights:")
print("1. Identify popular categories and brands in the dataset.")
print("2. Analyze the average price for common subcategories to see if certain items are generally more expensive.")
print("3. Observe which brands offer higher average discounts.")
print("4. Examine the most common product quantities available.")
print("5. Compare original and discounted prices for top brands to assess average discount levels.")
```
