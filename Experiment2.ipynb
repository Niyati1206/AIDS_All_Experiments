```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = Image.open("Data\img.jpg")

# Display the original image
plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis("off")
plt.title("Original Image")
plt.show()

# Step 1: Split the image into RGB channels
r, g, b = image.split()

# Display each color channel separately
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(r, cmap="Reds")
axes[0].set_title("Red Channel")
axes[0].axis("off")

axes[1].imshow(g, cmap="Greens")
axes[1].set_title("Green Channel")
axes[1].axis("off")

axes[2].imshow(b, cmap="Blues")
axes[2].set_title("Blue Channel")
axes[2].axis("off")
plt.show()

# Step 2: Plot histogram for each color channel
r_array = np.array(r).flatten()  # Convert red channel to a 1D array
g_array = np.array(g).flatten()  # Convert green channel to a 1D array
b_array = np.array(b).flatten()  # Convert blue channel to a 1D array

plt.figure(figsize=(10, 5))
plt.hist(r_array, bins=256, color="red", alpha=0.5, label="Red Channel")
plt.hist(g_array, bins=256, color="green", alpha=0.5, label="Green Channel")
plt.hist(b_array, bins=256, color="blue", alpha=0.5, label="Blue Channel")
plt.title("Color Intensity Histogram")
plt.xlabel("Intensity Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()

# Step 3: Display summary statistics for each color channel
print("Red Channel - Min:", r_array.min(), "Max:", r_array.max(), "Mean:", r_array.mean())
print("Green Channel - Min:", g_array.min(), "Max:", g_array.max(), "Mean:", g_array.mean())
print("Blue Channel - Min:", b_array.min(), "Max:", b_array.max(), "Mean:", b_array.mean())
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

# Specify the path to your audio file
file_path = 'Data\audio_sample.mp3'  # Replace with your actual audio file path

# Load audio file
y, sr = librosa.load(file_path, sr=None)  # Load the audio file

# Display waveform
plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform of Audio Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()

# Compute the spectrogram
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.figure(figsize=(10, 4))
librosa.display.specshow(D, sr=sr, x_axis="time", y_axis="log")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram of Audio Signal")
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Specify the path to your video file
video_path = 'video_sample.mp4'  # Change this to your actual video file path if necessary

# Read and display the video frames
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}. Please check the file path.")
else:
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        
        # Convert frame from BGR (OpenCV default) to RGB (for matplotlib)
        frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Display using matplotlib
        plt.imshow(frame_rgb)
        plt.axis('off')  # Hide axes
        plt.show(block=False)  # Non-blocking display
        plt.pause(0.03)  # Control playback speed

        # Optional: Add a delay to control the frame rate
        if cv2.waitKey(30) & 0xFF == ord('q'):  # Press 'q' to quit
            break

    cap.release()
    plt.close()

print("Video playback completed.")
```
